## <div id='matplotlib'>1. The matplotlib library</div>

Let's first import the matplotlib library:

````python
import matplotlib.pyplot as plt
````

### 1.A. How to make histograms?

Histograms can be produced using the ```hist()``` function. Let's see an example, let's generate numbers in the [0,10] range and make a histogram, if the number of generated numbers is sufficiently high, the distribution should be flat.

````python
from random import randrange

# Geneate 100000 random numbers b/w 0 and 10
n_to_generate = 100000
numbers = [randrange(10) for i in range(n_to_generate)]

# Plot the distribution of the generated numbers
plt.hist(numbers)
plt.show()
````

For more information and options see [matplotlib.pyplot.hist](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html).

### 1.B. How to make graphs?

````python
# Let's generate (x,y) points
x = [0, 1, 2, 3, 4, 5, 6]
y = [item * 2 for item in x]

# Let's plot y vs x
plt.plot(x, y)
plt.show()
````

For more information and options see [matplotlib.pyplot.plot](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html).

### 1.C. How to make scatter figures?

Let's generate some random data along two axes and make a scatter figure:

````python
# Generate data
from random import gauss
npoints = 1000
x = [gauss(5, 1) for i in range(npoints)]
y = [gauss(100, 10) for i in range(npoints)]

# Make a scatter figure
plt.scatter(x, y)
````

For more information and options see [matplotlib.pyplot.scatter](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html).

### 1.D. Styling your figures

#### Axis labels

For labeling the x-axis (y-axis), we use the ```xlabel()``` (```ylabel()```) function. Let's reproduce the histogram and add axis labels:

````python
from random import randrange

# Geneate 100000 random numbers b/w 0 and 10
n_to_generate = 100000
numbers = [randrange(10) for i in range(n_to_generate)]

# Plot the distribution of the generated numbers
plt.hist(numbers)
plt.xlabel('Randomly-generated number')
plt.ylabel('Entries')
plt.show()
````

#### Legends

Legends are short text descriptions of the curves/histograms/scatter plots shown. They are particular helpful when showing two or more histograms/graphs, so the user can easily identify what is being shown.

When creating a histogram, a graph or a scatter figure, we can add a label (a string that identifies it) by setting the ```label``` argument. Then, we can show all legends by using the ```legend()``` function. The position where the legend is shown is chosen automatically but we can choose a position ourselves by setting the ```loc``` attribute (options: ```'upper left'```, ```'upper right'```, ```'upper center'```, ```'lower left'```, ```'lower right'```, ```'lower center'```, ```'center left'```, ```'center right'```, ```'center'```).

Let's continue with our histogram produced before and add a legend to the lower left corner:

````python
from random import randrange

# Geneate 100000 random numbers b/w 0 and 10
n_to_generate = 100000
numbers = [randrange(10) for i in range(n_to_generate)]

# Plot the distribution of the generated numbers
plt.hist(numbers, label='Uniform distribution')
plt.xlabel('Randomly-generated number')
plt.ylabel('Entries')
plt.legend(loc='lower left')
plt.show()
````
